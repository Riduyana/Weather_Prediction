```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('/content/drive/MyDrive/Machine Learning/weather.csv')

print("Dataset preview: ")
df.head()

```

```python
df_1 = df[["MinTemp", "MaxTemp", "Rainfall","Humidity3pm"]]
df_1.head()
```

```python
avg_min_temp = np.mean(df["MinTemp"])
max_max_temp = np.max(df["MaxTemp"])
avg_rainfall = np.mean(df["Rainfall"])
avg_humidity = np.mean(df["Humidity3pm"])

print("\nAverage Min Temperature:", avg_min_temp)
print("Maximum Temperature:", max_max_temp)
print("Average Rainfall:", avg_rainfall)
print("Average Humidity:", avg_humidity)

# Plot the statistics
stats = [avg_min_temp, max_max_temp, avg_rainfall, avg_humidity]
labels = ["Avg Min Temp", "Average Max Temp", "Avg Rainfall", "Avg Humidity"]

plt.figure(figsize=(8,5))
plt.bar(labels, stats, color=['skyblue','orange','green','red'])
plt.title("Weather Data Statistics")
plt.ylabel("Values")
plt.show()

```
